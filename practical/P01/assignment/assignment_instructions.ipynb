## Assignment for practical session 01

The assignment consists of delivering a my_spiders.py file containing the "myFirstSpider()" class to scrape specific content from a particular URL. To achieve this, you need to modify the parse() method of the myFirstSpider() class accordingly. The "parse()" function should store the information inside the "my_dictionary" attribute of the myFirstSpider() class to be written to a dictionary file. This work is essentially what was done in the practical session, so you need to adapt the code to the specific requirement for the assignment. 

The URLs and information to scrape will be unique for each student, and therefore a different my_spiders.py file should be delivered by each student. The file should be submitted through the Moodle page of the course in the time limit there defined. 

Please, any doubts you have, do them through the forum so it can be useful to everyone.
